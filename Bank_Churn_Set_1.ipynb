### Part 2: Web Data Acquisition


```python
#Import Libraries
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# Extract banking dataset from a webpage
URL2 = "https://en.wikipedia.org/wiki/Banking_in_South_Africa"

headers = {"User-Agent": "Mozilla/5.0"}

wiki = requests.get(URL2, headers=headers)

if wiki.status_code == 200:
    banking_data_page = BeautifulSoup(wiki.content, "lxml")

    print(banking_data_page.title.text)

    # Extract all tables from the HTML and save into a dataframe called table_data
    table_data = pd.read_html(wiki.text)

    # Look at the third table (index 1)
    bank_tables = table_data[2]

    print("\nTable preview:")
    print(bank_tables.head(1))

else:
    print("Failed to retrieve page data!: code", wiki.status_code)

# Display the first five rows
print(f"\nFirst five rows of scraped dataframe:\n", bank_tables.head(5))
```
