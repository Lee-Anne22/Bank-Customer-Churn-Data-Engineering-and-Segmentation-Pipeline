#### Task 2: Consolidation

First, the merge() function is used when combining related data tables based on matching columns. In this project, the two dataframes have the same attributes and can use the CustomerId column as a primary key to join together, but the age and CreditScore attributes have different pandas data types. If we join the tables using merge(), selecting how="inner", then those two attribute columns (CreditScore and Age) will be kept individually and not merged with NaN values. If we select how="outer", the same outcome as how="inner" will occur. This will not provide the merged table we want and decrease data integrity by creating more NaN values and duplicates. If we want to use merge(), then the CreditScore and Age attributes will need to be converted to the same data type using astype(). However, even though this conversion could improve data integrity, there is a risk of truncation occuring with the float values becoming integers, this decreases precision. Also, the missing values (NaN's) in the model data set could fail with converting.

Next, the combine_first() function is another way to combine two data sets. It is used to combine data with the same set of entities but one data set may be missing data, this function fills in those gaps. However, for this project the goal is to combine the data sets into one long dataframe and not replace missing values with another source. This method will decrease the data integrity by filling any null-values with other customer data instead of that customers data. Hence, the accuracy of the data will be lowered and that is why this method will not be used. 

The last option would be to use concat(). Concat() is ideal for stacking data vertically, especially if is more records of the same structure. Due to the matching column records but different row numbers, stacking would work well. If we stack the churn data set (10000 rows) on top of the model data set (49 rows), we should get 10049 rows. It is similar to merge() due to also using joins but its advantage is that it aligns rows by index. The index can also be changed and the new data set can have observations (rows) with different variables(columns) because it automatically converts different data types into a compatible one for the whole column. Hence, to keep data intergity as much as possible, concat() will be used for this project.

NaN values are treated differently in NumPY compared to panda's. If a NaN value is used in a statistical calculation such as the mean, then the value remains a NaN and does not get treated as a zero. Also, any NaN values need to be dealt with to avoid inaccurate or biased results. Quantitative NaN values can be replaced by either the median or mean depending on if the data is symmetrical or skewed, therefore, a slight look into the current data using the describe() function will help.

```python
print(df_dup.describe())
```

According to the results, most of the columns with missing data is skewed to the left and one is skewed to the right. Therefore, the median should be used for these columns to replace the NaN values.

|Data column|Mean=Median?|Mean>Median|Mean<Median|Skewed|

The NaN values in the qualitative categories geography and gender will be replaced by a new label - "unknown". This is to keep the data integrity and avoid biased results. Even though the values missing is scarce compared to the dataframe size, it is better to handle the missing values conservatively to avoid the pitfalls mentioned.

```python
#Replacing NaN values 
def clean_missing_values(df_dup):
    num_cols = ['CreditScore', 'Age', 'Balance', 'Estimated Salary']
    cat_cols = ['Geography', 'Gender']
    
    
    df_dup[num_cols] = df_dup[num_cols].fillna(df_dup[num_cols].median())
    df_dup[cat_cols] = df_dup[cat_cols].fillna("Unknown")
    
    return df_dup
#Calling function
df = clean_missing_values(df_dup)

```

#### Task 6: Customer Segmentation and Value-Based Structuring
Customers at banks can be divided into various age bands according to what is being analysed. In this project, bank churn is being analysed, therefore, splitting customers into age bands based on life phases is implemented to see the churn during phases with different needs. 
The first phase would targert young adults who are transitioning from dependence on their parents to becoming independent by going to university or working. These are individuals who require their own banking for groceries, rent etc. and targets ages 18-30. 
The next phase includes adults who are working full time and are starting families or businesses. These individuals might require car loans or mortgages, financial planning advice or credit cards and targets 31-40 year olds.  
In the 41-50 age band are individuals whom are mid-career. These individuals may want more personalised banking options and start looking more into retirement options or investments. 
The 51-60 age band include individuals in the pre-retirement stage. This target audience may want more retirement options or to expand their invesment portfolios.
Followed by the retirement stage, individuals between 61-70 are banded together. People in this stage may look for alternative wealth management to optimise their retirement potential and may leave for better products.
Finally, individuals 71+ are considered retired and seniors. This group does not really change products due to the risks involved and may require advice for their will's.

Using the above as a guideline, these age bands will be used. 

```python
# Dividing customers in distinct age bands

# young_adults = df[(df['Age'] >= 18) & (df['Age'] <= 30)] # ages 18-30

# early_career = df[(df['Age'] >= 31) & (df['Age'] <= 40)] # ages 31-40

# mid_career = df[(df['Age'] >= 41) & (df['Age'] <= 50)] # ages 41-50

# pre_retire = df[(df['Age'] >= 51) & (df['Age'] <= 60)] # ages 51-60

# retired = df[(df['Age'] >= 61) & (df['Age'] <= 70)] # ages 61-70

# seniors = df[df['Age'] >= 71] # ages above 70

# Copilot Code Suggestion
def assign_age_band(age):
    if 18 <= age <= 30:
        return 'Young Adults (18–30)'
    elif 31 <= age <= 40:
        return 'Early Career (31–40)'
    elif 41 <= age <= 50:
        return 'Mid Career (41–50)'
    elif 51 <= age <= 60:
        return 'Pre-Retire (51–60)'
    elif 61 <= age <= 70:
        return 'Retired (61–70)'
    elif age >= 71:
        return 'Seniors (71+)'
    else:
        return 'Unknown'

df['AgeBand'] = df['Age'].apply(assign_age_band)

# print (young_adults) # testing code 
```

#### Task 6.2: Value base segmentation
Customer value to the bank can be categorised by looking at either the customer bank balance or estimated salary. For this project, the estimated salary will determine customer value. Customers will be categorised into a four level system based on what is considered below the mean, the lower mean, the higher mean and above the mean salary estimates for the current customer data base. The ranges for each category will be determined by the quartile ranges of the data, therefore, setting the standard on the bank customer demographic instead of restricting it to one region.

```python
# Finding the statistics on estimated salary
print(f"Summary Statistics for Estimated Salary:", df['Estimated Salary'].describe())
```

```python
#Value-Based Segmentation
# set variables
# min_val= df['Estimated Salary'].min()
# max_val= df['Estimated Salary'].max()
# mean_val= df['Estimated Salary'].mean()

# below_mean = df[(df['Estimated Salary']>= min_val) & (df['Estimated Salary']<= 0.25 * mean_val)]

# lower_mean = df[(df['Estimated Salary']>= 0.25 * mean_val) & (df['Estimated Salary']<= 0.50 * mean_val)]

# upper_mean = df[(df['Estimated Salary']>= 0.50 * mean_val) & (df['Estimated Salary']<= 0.75 * mean_val)]

# above_mean = df[(df['Estimated Salary']>= 0.75 * mean_val) & (df['Estimated Salary']<= max_val)]

# print(f"Below Mean Salary Estimate:", below_mean['New Customer Id'].count())
# print(f"Lower Mean Salary Estimate:", lower_mean['New Customer Id'].count())
# print(f"Upper Mean Salary Estimate:", upper_mean['New Customer Id'].count())
# print(f"Above Mean Salary Estimate:", above_mean['New Customer Id'].count())

# Copilot code suggestion (https://copilot.microsoft.com/shares/LCMxctZMvfL5WPMLQyJwq)
min_val = df['Estimated Salary'].min()
max_val = df['Estimated Salary'].max()
mean_val = df['Estimated Salary'].mean()

def assign_salary_band(sal):
    if min_val <= sal <= 0.25 * mean_val:
        return 'Below Mean (0–25%)'
    elif 0.25 * mean_val < sal <= 0.50 * mean_val:
        return 'Lower Mean (25–50%)'
    elif 0.50 * mean_val < sal <= 0.75 * mean_val:
        return 'Upper Mean (50–75%)'
    elif 0.75 * mean_val < sal <= max_val:
        return 'Above Mean (75–100%)'
    else:
        return 'Unknown'

df['SalaryBand'] = df['Estimated Salary'].apply(assign_salary_band)
```

#### Task 7: Grouped Aggregation Analysis
The following aggregate questions were answered using code from Copilot (https://copilot.microsoft.com/shares/LCMxctZMvfL5WPMLQyJwq) :
1. Mean Estimated Salary by Age Group - To compare financial increases in different life phases.

```python
# Mean Estimated Salary by Age Group
salary_by_age = df.groupby('AgeBand')['Estimated Salary'].mean()
print (f"Mean Estimated Salary by Age Group\n", salary_by_age)
# Mean CreditScore by Churn Status
avg_creditscore_by_churn_status = df.groupby('Churn')['CreditScore'].mean()
print(f"\nMean Credit Score by Churn Status\n 0 - No\n 1 - Yes\n", avg_creditscore_by_churn_status)
# Churn rate by Geography, Gender and Age group
churn_rate_via_GGA = df.groupby(['Geography', 'Gender', 'AgeBand'])['Churn'].mean()
print(f"\n Churn Rate by Geography, Gender and Age Band\n", churn_rate_via_GGA )
# Churn rate by number of products and Estimated Salary
churn_rate_by_products = df.groupby(['Num Of Products', 'SalaryBand'])['Churn'].mean()
print (f"\nChurn Rate by Number of Products and Estimated Salary\n", churn_rate_by_products)
# Mean Creditscore by Age band and Gender
mean_credit_by_age_gender = df.groupby(['AgeBand', 'Gender'])['CreditScore'].mean()
print (f"\nMean Credit Score by Age Band and Gender\n", mean_credit_by_age_gender)
# Mean Member Activity by Geography and Age
mean_ma = df.groupby(['Geography', 'AgeBand'])['Is Active Member'].mean()
print(f"\n Mean Member Activity by Geography and Age\n", mean_ma)
# Mean Age and balance by Geography
mean_age_and_balance = df.groupby('Geography')[['Age', 'Balance']].mean()
print(f"\nMean Age and Balance by Geography\n",mean_age_and_balance)

# Copilot code help: (https://copilot.microsoft.com/shares/LCMxctZMvfL5WPMLQyJwq)

```
